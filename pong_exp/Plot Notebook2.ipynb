```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os.path
```

```python
EXPERIMENTS = [
    '10K_lr=0.1_gamma=0.5_rnd_prob=0.5',
    '10K_lr=0.1_gamma=0.7_rnd_prob=0.5',
    '10K_lr=0.1_gamma=0.9_rnd_prob=0.5',
    '10K_lr=0.3_gamma=0.5_rnd_prob=0.5',
    '10K_lr=0.3_gamma=0.7_rnd_prob=0.5',
    '10K_lr=0.3_gamma=0.9_rnd_prob=0.5',
    '10K_lr=0.7_gamma=0.5_rnd_prob=0.5',
    '10K_lr=0.7_gamma=0.9_rnd_prob=0.5',
    '20K_lr=0.7_gamma=0.7_rnd_prob=0.5',
]

METRICS = [
    'pong Reward per episode (smooth)',
    'pong Average paddle bounces per point (smooth)',
    'pong Cumulative Reward'
]

def plot_results(metrics, experiments):
    for metric in metrics:
        fig, ax = plt.subplots(figsize=(25, 5))
        for exp in experiments:
            df = pd.read_csv(os.path.join('/home/lfx/Desktop/Netlogo-pong/', exp , metric) + '.csv', skiprows=16, sep=',')
            if 'smooth' in metric:
                df['x'] = df['x']*50
            plt.xlim([0, 10000])

            if metric == 'pong Reward per episode (smooth)':
                plt.ylim([-21, 21])
            plt.ylabel(metric)
            plt.xlabel('episodes (smooth)')
            sns.lineplot(x=df['x'], y=df['y'], ax=ax, label=exp, legend=True)
```

# Learning Rate

```python
plot_results([METRICS[0]], [exp for exp in EXPERIMENTS if 'lr=0.1' in exp])
```

```python
plot_results([METRICS[0]], [exp for exp in EXPERIMENTS if 'lr=0.3' in exp])
```

```python
plot_results([METRICS[0]], [exp for exp in EXPERIMENTS if 'lr=0.7' in exp])
```
